```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.regime_switching.tests.test_markov_regression import fedfunds, ogap, fedfunds_const_filtered_joint_probabilities
from markov_switch import MarkovSwitch
```

```python
np.set_printoptions(suppress=True)
msm = MarkovSwitch()
msm.fit(np.array(ogap))
```

```python
plt.figure(figsize=(14,5))
plt.title("ogap time series")
plt.plot(ogap)
plt.show()
```

```python
plt.figure(figsize=(14,5))
dates = pd.date_range('1954-07-01', '2010-10-01', freq='QS')
plt.plot(dates, msm.filtered_p[:,0], label="$P(S_t = 0 | F_t)$")
plt.plot(dates, msm.filtered_p[:,1], label="$P(S_t = 1 | F_t)$", color='r', linestyle='--')
plt.legend(loc='upper left') # displays labels
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(fedfunds)
plt.show()
```
